```python
# 3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Создаем словарь с данными о продажах (с пропусками и выбросами)
data = {
    'Дата': pd.date_range(start='2024-01-01', periods=20, freq='D'),
    'Товар': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C',
              'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Количество': [10, 15, 8, 12, 18, 9, 14, 20, 7, np.nan,
                  16, 11, 13, 250, 6, 15, 19, 8, np.nan, 22],
    'Цена': [100, 200, 150, 300, 180, 120, 280, 190, 130, 320,
             210, 140, 290, 50, 110, 310, 220, 160, np.nan, 240],
    'Регион': ['Север', 'Юг', 'Восток', 'Запад', 'Север', 'Юг', 'Восток', 'Запад',
               'Север', 'Юг', 'Восток', 'Запад', 'Север', 'Юг', 'Восток', 'Запад',
               'Север', 'Юг', 'Восток', 'Запад']
}


df = pd.DataFrame(data)
median_column = df['Цена'].median()
fi = df['Цена'].fillna(median_column, inplace = True)

outliers = df[(df['Количество'] < 1) | (df['Количество'] > 1000)]
initial_count = len(df)
df_cleaned = df[(df['Количество'] >= 1) & (df['Количество'] <= 1000)]
removed_count = initial_count - len(df_cleaned)
new = df['Общая_стоимость' ] = df['Цена'] * df['Количество']
ol = df.groupby('Товар')['Общая_стоимость'].sum().sort_values(ascending=False)

print("Пропуски в столбце 'Цена' заполнены медианным значением:")
print(f"Медианная цена: {median_column:.2f}")
print("\nПроверка пропущенных значений после заполнения:")
print(df.isnull().sum())
print(f"\nНайдено строк с выбросами: {len(outliers)}")
print("Строки с выбросами:")
print(outliers)
print(f"\nУдалено строк: {removed_count}")
print(f"Осталось строк: {len(df_cleaned)}")
print(new)
print(ol, 'Общая_стоимость')
plt.figure(figsize=(10, 6))
bars = plt.bar(ol.index, ol.values,
               color=['skyblue', 'lightgreen', 'lightcoral'],
               edgecolor='black', linewidth=1.2)

# Настраиваем внешний вид графика
plt.title('Суммарная выручка по товарам', fontsize=16, fontweight='bold', pad=20) # pad = отступ от графика
plt.xlabel('Товар', fontsize=12, fontweight='bold') #fontsize = размер шрифта
plt.ylabel('Выручка', fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--') #alpha = прозрачность
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
```

```python
# 4
import numpy as np
import numpy as np
A = np.random.randint(1, 11, size=(5, 5))
B = np.random.randint(1, 11, size=(5, 5))
print("Матрица A (5x5):")
print(A)
print("\nМатрица B (5x5):")
print(B)
sum_product = A * B
print("\nПоэлементное произведение матриц:")
print(sum_product)
multi_mat = np.dot(A, B)
print("\nМатричное произведение:")
print(multi_mat)
det_A = np.linalg.det(A)
print(f"\nОпределитель матрицы A: {det_A:.2f}")
A_inv = np.linalg.inv(A)
print("\nОбратная матрица A⁻¹:")
print(A_inv)
C = A.sum(axis=1).reshape(-1, 1)

print("\nВектор C (суммы строк матрицы A):")
print(C)
x = np.linalg.solve(A, C)
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

fig.colorbar(surf, ax=ax, shrink=0.6, aspect=20)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$z = \sin\left(\sqrt{x^2 + y^2}\right)$')

plt.show()
```
